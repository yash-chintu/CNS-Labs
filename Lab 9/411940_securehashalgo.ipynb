# Secure Hash Algo

```python
def complement(a,b,c):
    return((a&b)^(a&c)^(b&c))
```

```python
def XOR(a,b,c):
    return(a^b^c)
```

```python
def rotateleft(x,n,w):
    return((x<<n&(2**w-1))|(x>>w-n))
```

```python
def And(a,b,c):
    return((a&b)^(~a&c))
```

```python
def SHA_1(x):
    key = []
    for t in range(80):
        if   t <= 19:
            key.append(0x5a827999)
        elif t <= 39:
            key.append(0x6ed9eba1)
        elif t <= 59:
            key.append(0x8f1bbcdc)
        else:
            key.append(0xca62c1d6)
    x_bytes = bytearray(x, 'ascii')
    x_bits  = [format(x, '08b') for x in x_bytes]
    #print('x_bits_string:', x_bits)
    x_bits_string = ''.join(x_bits)
    print('x_bits_string:', x_bits_string)
    pad_bits = '1' + ('0' * (448 - (8 * len(x) + 1))) + format(len(x) * 8, '064b')
    x_padded = x_bits_string + pad_bits
    print('x_padded_string:', x_padded)
    #print("It's length is :", len(x_padded))
    assert(len(x_padded) == 512)
    M1 = x_padded
    Hash = [0x67452301, 0xefcdab89, 0x98badcfe, 0x10325476, 0xc3d2e1f0]
    N = 1
    for i in range(1, N + 1):
        print('' * 2)
        print('i = ', i)
        W = list()
        for t in range(80):
                if t <= 15:
                    W.extend([ int(M1[ (32 * t) : (32 * (t + 1)) ], 2)  ])
                else:
                    W.extend([ rotateleft( W[t - 3] ^ W[t - 8] ^ W[t - 14] ^ W[t - 16], n=1, w=32) ])
        print('W:', W[0:16])
        a = Hash[0]
        b = Hash[1]
        c = Hash[2]
        d = Hash[3]
        e = Hash[4]
        print('hexadecimal1:', hex(a))
        print('hexadecimal2:', hex(b))
        print('hexadecimal3:', hex(c))
        print('hexadecimal4:', hex(d))
        print('hexadecimal5:', hex(e))
    for t in range(80):
        print('')
        print('t =', t)
        if t <= 19:
            f = And
        elif t <= 39:
            f = XOR
        elif t <= 59:
            f = complement
        else:
            f = XOR
        T = (rotateleft(a, n=5, w=32) + f(b, c, d) + e + key[t] + W[t]) % (2 ** 32)
        e = d
        d = c
        c = rotateleft(b, n=30, w=32)
        b = a
        a = T
        print('hexadecimal1:', hex(a))
        print('hexadecimal2:', hex(b))
        print('hexadecimal3:', hex(c))
        print('hexadecimal4:', hex(d))
        print('hexadecimal5:', hex(e))
    Hash[0] = (a + Hash[0]) % (2 ** 32)
    Hash[1] = (b + Hash[1]) % (2 ** 32)
    Hash[2] = (c + Hash[2]) % (2 ** 32)
    Hash[3] = (d + Hash[3]) % (2 ** 32)
    Hash[4] = (e + Hash[4]) % (2 ** 32)
    print("Message digest: ",Hash)
    Hash = [format(j, '08x') for j in Hash]
    return("".join(Hash))
```
